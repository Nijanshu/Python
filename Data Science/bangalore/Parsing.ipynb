```python
import json
```

```python
with open("finaldata.txt", encoding="utf-8") as f:
    data=f.read()

data
chunks= data.split("\n\n")
chunks=[c for c in chunks if len(c)>3]
chunks[10]
```

```python
def parse_chunk(chunk):
    chunk=chunk.strip()
    lines=chunk.split("\n")
    username=lines[0]
    posts=int(lines[1].split(" ")[0].replace(",", ""))
    followers=float(lines[2].split(" ")[0].replace(",","").replace("K","").replace("M", ""))
    following=float(lines[3].split(" ")[0].replace(",","").replace("K","").replace("M", ""))

    if("K" in lines[2]):
            followers = int(followers * 1000)
    elif("M" in lines[2]):
            followers = int(followers * 1000000)
    else:
            followers = int(following)
      

    if("K" in lines[3]):
            following = int( following* 1000)
    elif("M" in lines[3]):
            following = int( following* 1000000)
    else:
            following = int(following)


    name=lines[4]
    if(len(lines)> 5):
         type_of_page = lines[5]
         bio = "\n".join(lines[6:])
    else:
         type_of_page = "Unknown"
         bio = "" 

    return {"username": username, "no_of_posts": posts, "no_of_followers": followers,
                "no_of_following": following, "name": name, "type_of_page": type_of_page, "bio": bio}
    
    
```

```python
all_chunks = [parse_chunk(c) for c in chunks] 

s=json.dumps(all_chunks, indent=4)
print(all_chunks[4])
with open("data2.json", "w") as f:
    f.write(s)
```
